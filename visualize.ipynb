```python
import os
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for the bar graph
intents = []
questions = []
answers = []

with open('./Intent.json') as dataset:
    dataset = json.loads(dataset.read())
    for intent in dataset['intents']:
        intents.append(intent["intent"])
        questions.append(len(intent["text"]))
        answers.append(len(intent["responses"]))
```

```python
# Create the bar graph
plt.figure(figsize=(12,6)) 
barWidth = 0.4
r1 = np.arange(len(questions))
r2 = [x + barWidth for x in r1]

plt.bar(r1, questions, color='#0099ff', width=barWidth, edgecolor='white', label='Questions Number')
plt.bar(r2, answers, color='#009933', width=barWidth, edgecolor='white', label='Answers Number')

# Add labels and title
plt.xlabel("Intent Name")
plt.ylabel("Number of Q/A")
plt.title("Number of Q/A per Intent")
plt.xticks([r + barWidth/2 for r in range(len(questions))], intents, rotation=45, fontsize=8)
plt.legend()

# Show the graph
plt.show()
```
